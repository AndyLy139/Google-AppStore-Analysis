```python
od.download("https://www.kaggle.com/datasets/gauthamp10/google-playstore-apps")
```

```python
original_df = pd.read_csv('/Users/andyb/google-playstore-apps/Google-Playstore.csv')

```

```python
original_df     # has 2.3 million rows wit lots of NaNs (Null Values)

cleaned_df=original_df.dropna() # drop na removes any row with Nans
cleaned_df # now only has 1.2 million rows
```

```python
cleaned_df = cleaned_df.drop(['Minimum Installs',
                             'Maximum Installs',
                              'Price',
                             'Developer Website',
                             'Developer Email', 
                             'Privacy Policy',
                             'Editors Choice'], axis =1
                            )
```

```python
cleaned_df['Installs'] = cleaned_df['Installs'].str.replace('+', '').str.replace(',', '').astype(int)
cleaned_df
```

```python
#Step 3: Statistical Analysis
#• Count(s)
#• Min, max, range
#• Standard deviation
#• Distribution
#• Outliers
#• Correlation (if merging data)
#• What are some interesting statistics about your data?
```

```python
num_content_ratings = cleaned_df['Content Rating'].nunique()
print("Number of unique content ratings: ", num_content_ratings)

# count the number of unique content ratings and sort them alphabetically
content_ratings = cleaned_df["Content Rating"].value_counts().sort_index()
# print the content ratings and their counts
for rating, count in content_ratings.items():
    print(f"{rating}: {count}")
```

```python
# set the float format
pd.options.display.float_format = '{:,.2f}'.format

# compute the summary statistics for the 'Installs' column
stat = cleaned_df['Installs'].describe()
```
